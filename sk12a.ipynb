```python
import pandas as pd

# 1. Load the data
try:
    df = pd.read_csv("C:/Users/muthuram/Downloads/student_scores.csv")
except FileNotFoundError:
    print("Error: 'students_scores.csv' not found. Please ensure the file exists.")
    exit()

# --- 1. Overall Class Performance ---
average_score = df['Score'].mean()
highest_score = df['Score'].max()
lowest_score = df['Score'].min()

print("--- Overall Class Performance ---")
print(f"Average: {average_score:.2f}")
print(f"Highest: {highest_score}")
print(f"Lowest:  {lowest_score}\n")

# --- 2. Pass/Fail Classification (Pass mark is 50) ---
# We create a new column 'Status' based on the 'Score'
df['Status'] = df['Score'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')

# --- 3. Grading on a Curve/Thresholds ---
def assign_grade(score):
    if score >= 90: return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60: return 'D'
    else: return 'F'

df['Grade'] = df['Score'].apply(assign_grade)

# Display the final analysis table
print("--- Student Analysis ---")
print(df[['Student_Name', 'Score', 'Status', 'Grade']])

# Optional: Save the analysis to a new CSV
# df.to_csv('analyzed_student_scores.csv', index=False)
```
