```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load the dataset and display column names/data types ---
# Using the data provided in the table
data = {
    'Patient_ID': ['D001', 'D002', 'D003', 'D004', 'D005', 'D006', 'D007', 'D008', 'D009', 'D010'],
    'Name': ['Alice', 'Bob', 'Clara', 'David', 'Ella', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack'],
    'Age': [54, 61, 47, 59, 65, 50, 56, 62, 60, 58],
    'Gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M'],
    'Diagnosis': ['Diabetes', 'Diabetes', 'Diabetes', 'Diabetes', 'Diabetes', 'Diabetes', 'Diabetes', 'Diabetes', 'Diabetes', 'Diabetes'],
    'Last_Visit': ['2024-07-10', '2024-07-05', '2024-06-30', '2024-07-15', '2024-07-01', '2024-07-08', '2024-06-28', '2024-07-12', '2024-07-14', '2024-07-06'],
    'Medication': ['Metformin', 'Insulin', 'Glipizide', 'Metformin', 'Insulin', 'Metformin', 'Glimepiride', 'Insulin', 'Metformin', 'Glipizide'],
    'HbA1c': [7.5, 8.2, 6.8, 7.9, 8.4, 7.1, 7.6, 8.0, 7.3, 7.7],
    'Glucose_Level': [180, 210, 160, 195, 220, 175, 185, 200, 170, 190]
}

df = pd.DataFrame(data)

print("--- 1. Column Names and Data Types ---")
print(df.info()) # Displays columns, non-null counts, and dtypes
print("\n")

# --- 2. Calculate and display the average age of the patients ---
average_age = df['Age'].mean()
print(f"--- 2. Statistical Analysis ---")
print(f"Average Patient Age: {average_age:.1f} years\n")

# --- 3. Analyze the distribution of diagnoses ---
diagnosis_counts = df['Diagnosis'].value_counts()
print("--- 3. Most Common Medical Conditions ---")
print(diagnosis_counts)
print("\n")

# --- 4. Visualize the diagnosis distribution (and Glucose Levels) ---
# Set the visual style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Subplot 1: Count Plot for Medication (since Diagnosis is all the same in this set)
plt.subplot(1, 2, 1)
sns.countplot(x='Medication', data=df, hue='Medication', palette='magma', legend=False)
plt.title('Distribution of Medications Prescribed')

# Subplot 2: Scatter plot of Age vs Glucose Level
plt.subplot(1, 2, 2)
sns.scatterplot(x='Age', y='Glucose_Level', size='HbA1c', data=df, color='teal', sizes=(50, 200))
plt.title('Patient Age vs. Glucose Levels (Size = HbA1c)')

plt.tight_layout()
plt.show()
```
