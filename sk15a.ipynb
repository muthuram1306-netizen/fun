```python
import pandas as pd

# 1. Load and organize the EEG data
try:
    df = pd.read_csv("C:/Users/muthuram/Downloads/eeg_data.csv")
    # Setting 'Time' as index helps in identifying patterns over time
    df.set_index('Time', inplace=True)
    print("--- EEG Data Loaded Successfully ---\n")
    print(df.head(), "\n")
except FileNotFoundError:
    print("Error: 'eeg_data.csv' not found.")
    exit()

# 2. Perform basic statistical analysis
# We calculate Mean, Variance, and Peak (Max) values for each electrode
stats = pd.DataFrame({
    'Mean': df.mean(),
    'Variance': df.var(),
    'Peak Value': df.max(),
    'Minimum': df.min()
})

print("--- Statistical Analysis per Electrode ---")
print(stats, "\n")

# 3. Identify patterns or trends
print("--- Pattern Analysis ---")

# Trend: Correlation between electrodes
correlation = df.corr()
print("Correlation Matrix (Checking if brain regions fire together):")
print(correlation, "\n")

# Identifying specific activity peaks
for electrode in df.columns:
    peak_time = df[electrode].idxmax()
    peak_val = df[electrode].max()
    print(f"Electrode {electrode} reached peak activity of {peak_val:.2f} µV at {peak_time}s.")
```
