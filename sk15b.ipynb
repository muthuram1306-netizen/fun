```python
import pandas as pd
import matplotlib.pyplot as plt

# The provided dictionary
data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

print("--- Initial Inventory Data ---")
print(df, "\n")

# --- OPERATION 1: Calculate Total Inventory Value per Product ---
df['Total_Value'] = df['Price'] * df['Quantity']

# --- OPERATION 2: Basic Descriptive Statistics ---
# Provides mean, min, max, etc., for numerical columns
stats = df.describe()

# --- OPERATION 3: Filtering (Products with Price > 200) ---
expensive_products = df[df['Price'] > 200]

# --- OPERATION 4: Sorting (By Quantity Descending) ---
sorted_inventory = df.sort_values(by='Quantity', ascending=False)

# --- OPERATION 5: Calculate Grand Total Value of all Stock ---
grand_total = df['Total_Value'].sum()

# Display Results
print("--- Analysis Results ---")
print(f"1. Total Value per Product:\n{df[['Product', 'Total_Value']]}\n")
print(f"2. Statistical Summary:\n{stats}\n")
print(f"3. Products priced above 200:\n{expensive_products}\n")
print(f"4. Inventory sorted by Quantity:\n{sorted_inventory}\n")
print(f"5. Total Warehouse Stock Value: ${grand_total}\n")

# --- PLOTTING ---
plt.figure(figsize=(10, 5))

# Subplot 1: Bar Chart of Quantities
plt.subplot(1, 2, 1)
plt.bar(df['Product'], df['Quantity'], color='skyblue')
plt.title('Stock Quantity by Product')
plt.xlabel('Product')
plt.ylabel('Quantity')

# Subplot 2: Pie Chart of Total Value Distribution
plt.subplot(1, 2, 2)
plt.pie(df['Total_Value'], labels=df['Product'], autopct='%1.1f%%', startangle=140)
plt.title('Total Value Distribution')

plt.tight_layout()
plt.show()
```
