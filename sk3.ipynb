```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("C:/Users/muthuram/Downloads/flight_data.csv")

print("--- Flight Data Analysis ---")

# a. Display summary statistics for departure delays
# This gives us count, mean, std, min, 25%, 50% (median), 75%, and max
print("\n[A] Summary Statistics for Departure Delays:")
print(df['departure_delay'].describe())

# b. Calculate the average delay for each airline
print("\n[B] Average Delay by Airline:")
avg_delay = df.groupby('airline')['departure_delay'].mean()
print(avg_delay)

# c. Identify the most common departure and arrival destinations
common_origin = df['origin'].mode()[0]
common_dest = df['destination'].mode()[0]
print(f"\n[C] Most Common Departure: {common_origin}")
print(f"    Most Common Arrival: {common_dest}")

# d. Visualize the distribution of departure delays using a box plot
print("\n[D] Generating Box Plot...")

plt.figure(figsize=(8, 6))
plt.boxplot(df['departure_delay'], patch_artist=True, labels=['Departure Delay'])

plt.title('Distribution of Departure Delays')
plt.ylabel('Minutes')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
